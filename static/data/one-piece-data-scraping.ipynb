```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
URL = "https://www.sportskeeda.com/anime/how-many-devil-fruits-one-piece-all-205-fruits-listed"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup)
```

```python
results = soup.find(class_="content-holder")
```

```python
print(results.prettify())
```

```python
list_elements = results.find_all("li")
print(list_elements)
```

```python
df = pd.DataFrame(columns=["Fruit", "Ability Description", "Name", "Fruit Type"])
rows = []

for list_element in list_elements:
    list_splits = list_element.text.split(" - ")
    fruit = list_splits[0]
    ability = list_splits[1].split(" User: ")[0]
    user = list_splits[1].split(": ")[1] if len(list_splits[1].split(": ")) > 1 else None
    
    rows.append({"Fruit": fruit, "Ability Description": ability, "Name": user})
    
df = pd.concat([df, pd.DataFrame(rows)], ignore_index=True)
```

```python
df['Name'] = df['Name'].str.strip(".")
```

```python
df['Fruit Type'][0:88] = 'Paramecia'
df['Fruit Type'][88:152] = 'Zoan'
df['Fruit Type'][152:164] = 'Logia'
df['Fruit Type'][164:] = 'Non-Canon Fruit'
```

```python
# bounty data scraping
URL2 = "https://onepiece.fandom.com/wiki/Bounties/List"
page2 = requests.get(URL2)

soup2 = BeautifulSoup(page2.content, "html.parser")
print(soup2)
```

```python
dfs = pd.read_html(URL2)
```

```python
dfs[34]
```

```python
df3 = pd.DataFrame(columns=["Name", "Bounty", "Affiliation", "Status"])

df3['Name'] = dfs[34]['Name']
df3['Bounty'] = dfs[34]['Bounty']
df3['Affiliation'] = dfs[34]['Affiliation']
df3['Status'] = dfs[34]['Status']

print(df3.to_string())
```

```python
final = pd.merge(df, df3, how="left", on='Name')
final['Ability Description'] = final['Ability Description'].apply(lambda x: f'"{x}"')
final.fillna(value="Unknown", inplace=True)
print(final.head())
```
